1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:

2. Graficar los siguientes grupos de funciones:

Grupo 1:
    -$$ y_{1} = 2x+1 $$
    -$$ y_{2} = -x+4 $$
    -$$ y_{3} = x^2 $$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10)
y1 = 2*x +1
y2 = -x + 4
y3 = x**2
plt.figure()
plt.grid()
plt.plot(x,y1, "r", label="primera ecuación")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("GRUPO 1")
plt.plot(x, y2, "g", label="segunda ecuación")
plt.plot(x, y3, "b", label="tercera ecuación")
plt.legend(loc = "best")  # simbologia
```

Grupo 2:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10) 
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()
plt.grid()
plt.plot(x,y1, "r", label="primera ecuación")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("GRUPO 2")
plt.plot(x, y2, "g", label="segunda ecuación")
plt.legend(loc = "best")  # simbologia
```

3. Crea una función que cree una gráfica y dibuje 'k' líneas de forma $$f(x)=sin(\alpha x)$$

La función recibe:
- x1: valor inicial de 'x'
- x2: valor final de 'x
- n: cantidad de puntos del vector 'x'
- a: valor inicial $ \alpha$
- b: valor final $ \alpha$
- k: cantidad de líneas

```python
def funcion(x1,x2,n,a,b,k):
    import matplotlib.pyplot as plt
    import numpy as np
    
    # x1 = 0 # x2 = 2*np.pi # n= 1000
    x = np.linspace(x1, x2,n)
    # a = 1 # b = 3 # k = 3
    alpha= np.linspace(a, b,k)
    plt.figure(figsize=(8,6))
    for a in alpha :
        y = np.sin(a * x)
        plt.plot(x,y, label="línea alpha " + str(a)) #plt.plot(x , y)

    
    plt.grid()
    plt.xlabel("Eje x")
    plt.ylabel("Eje y")
    plt.title("FUNCION PARA LINEAS")
    plt.legend(loc = "best")  # simbologia
```

```python
import numpy as np

funcion(0,2*np.pi,1000,1,3,3)
```
